```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from datetime import datetime
```

```python
stations = np.array(['FPZ_06A', 'FSB_079', 'FSB_095DR', 'FSB_078', 'FSB_130D',
       'FEX_004', 'FSB_126', 'FPZ_04A', 'FSB_097D', 'FAI_012', 'FAI_009'])
```

```python
index = 10
df = pd.read_csv(f"../data/merged_data_stationwise/{stations[10]}.csv")
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['Date Time'] = df['Date Time'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S').timestamp() * 1000)
df = df.drop('station_id', axis=1)
```

```python
df.head()
```

```python
column_name = df.columns
```

```python
num_to_name = {}
for index, val in enumerate(column_name):
    num_to_name[index] = val
```

```python
imputer = KNNImputer(n_neighbors = 5)
df_imputed_scaled = imputer.fit_transform(df)
df = pd.DataFrame(df_imputed_scaled)
```

```python
df.head()
```

```python
df = df.rename(columns = num_to_name)
df.head()
```

```python
df.isna().sum()
```

```python
df2 = df.copy()
scaler = StandardScaler()
df2[df2.columns] = scaler.fit_transform(df2[df2.columns])
```
