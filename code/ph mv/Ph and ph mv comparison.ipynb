```python
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as  plt
import plotly.graph_objects as go
```

```python
# all the unique station in the provided dataset
unique_stations = ['FPZ_04A', 'FSB_126', 'FSB_079', 'FPZ_06A', 'FEX_004', 'FSB_078',
       'FSB_097D', 'FSB_130D', 'FSB_095DR']
finder = {}
for index, val in enumerate(unique_stations):
    finder[val]= index

finder
```

```python
df = []
for i in range(9):
    temp = pd.read_csv(f"./data/{unique_stations[i]}.csv")
    df.append(temp)
```

```python
plt.figure(figsize=(10, 5))
colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'purple','turquoise','lime', 'gold', "slategray"]
for i in range(9):
    temp = df[i]
    plt.plot(temp["pH (pH)"], temp["pH MV (mV)"], marker='.', linestyle='-', color=colors[i])

plt.xlabel('pH (pH)')
plt.ylabel('pH MV (mV)')
plt.title(f'Plot of pH over pH MV (mV).')
plt.grid(True)
plt.show()
```
